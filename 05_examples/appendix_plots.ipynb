```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# PLOTS FOR SIMPLE MODELS (MINMAX SCALING)
# Data preparation
df = pd.DataFrame({
    'Model': ['LightGBM'] * 4 + ['RNN-GRU'] * 4 + ['TFT'] * 4,
    'Type': ['Simple', 'Simple CQR', 'Covariates', 'Covariates CQR'] * 3,
    'Mean Width': [5.7036, 6.8776, 6.132, 6.089,
                   5.1256, 5.7203, 4.6918, 6.7249,
                   5.8443, 6.272, 4.3975, 5.709],
    'Coverage': [0.8357, 0.9137, 0.8512, 0.9089,
                 0.8673, 0.9012, 0.85, 0.8872,
                 0.8393, 0.908, 0.8537, 0.9131]
})

df['Method'] = df['Model'] + ' - ' + df['Type']

# Apply consistent plotting style
sns.set(style="whitegrid", font_scale=1.0)

# Function to add value labels to bars
def add_labels(ax, values, offset=0.01, fmt="{:.2f}"):
    for i, v in enumerate(values):
        ax.text(v + offset, i, fmt.format(v), va='center')

# Plot 1: Mean Width
plt.figure(figsize=(10, 6))
df_sorted = df.sort_values('Mean Width')
ax1 = sns.barplot(data=df_sorted, x='Mean Width', y='Method', hue='Model', dodge=False, palette='Blues')
add_labels(ax1, df_sorted['Mean Width'], offset=0.05)
plt.xlim(2, 8)
plt.title("Mean Prediction Interval Width by Method")
plt.xlabel("Mean Width (L/day)")
plt.ylabel("")
plt.legend(title='Model', loc='upper right')
plt.tight_layout()

filename = f"/content/drive/MyDrive/Colab Notebooks/TFG/05_examples/minmax_comparison_width.png"
plt.savefig(filename, dpi=300)

plt.show()

# Plot 2: Coverage
plt.figure(figsize=(10, 6))
df_sorted = df.sort_values('Coverage')
ax2 = sns.barplot(data=df_sorted, x='Coverage', y='Method', hue='Model', dodge=False, palette='Greens')
add_labels(ax2, df_sorted['Coverage'], offset=0.003, fmt="{:.3f}")
plt.axvline(0.90, color='red', linestyle='--', label='Goal (0.9)')
plt.xlim(0.8, 1.0)
plt.title("Prediction Interval Coverage by Method")
plt.xlabel("Coverage")
plt.ylabel("")
plt.legend(title='Model', loc='lower right')
plt.tight_layout()

filename = f"/content/drive/MyDrive/Colab Notebooks/TFG/05_examples/minmax_comparison.png"
plt.savefig(filename, dpi=300)

plt.show()
```
