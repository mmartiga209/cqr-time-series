# DAILY DATA
## First dataset
Within the daily_consumption dataset we have data divided in two csv files. Starting with the daily_consumption:

```python
df_daily_census = pd.read_csv("../data/datasets/daily_consumption_census.csv", low_memory=False)
```

```python
print("DATA AGGREGATED BY CENSUS SECTION:")
df_daily_census
```

```python
# These are the columns of the dataset
columns_to_check = [
    "Census section",
    "District",
    "Municipality",
    "Date",
    "Use",
    "Number of meters",
    "Accumulated consumption (L/day)"
]

# We will go thorugh all columns to check the different values they get
for col in columns_to_check:
    values = df_daily_census[col].replace('<NULL>', pd.NA).dropna().unique()    # We have to remove the "<NULL>" values
    print(f"COLUMN: {col}") 
    print(f"Unique values: {len(values)}")
    print(f"Values: {values}\n")
```
